```python
# Gaussian Filtering
import cv2
import numpy as np
from matplotlib import pyplot as plt
img = cv2.imread('sun.jpg')

# Change color to RGB (from BGR)
image = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Convert the image to grayscale for processing
gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)

# gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
blur = cv2.GaussianBlur(gray,(3,3),1) # 3,3 kernel


plt.subplot(121)
plt.imshow(gray, cmap='gray')
plt.title('Original')
plt.xticks([]), plt.yticks([])

plt.subplot(122)
plt.imshow(blur, cmap='gray')
plt.title('Blurred')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
# Gaussian Filtering
import cv2
import numpy as np
from matplotlib import pyplot as plt
img = cv2.imread('download.jpg')

# Change color to RGB (from BGR)
image = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Convert the image to grayscale for processing
gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)

# gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
blur = cv2.GaussianBlur(image,(3,3),4) # 3,3 kernel


plt.subplot(121)
plt.imshow(image)
plt.title('Original')
plt.xticks([]), plt.yticks([])

plt.subplot(122)
plt.imshow(blur, cmap='gray')
plt.title('Blurred')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
# Median Filtering
import cv2
import numpy as np
from matplotlib import pyplot as plt
img = cv2.imread('sun.jpg')
# Change color to RGB (from BGR)
image = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Convert the image to grayscale for processing
gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)

# gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
# blur = cv2.GaussianBlur(gray,(5,5),0)
median = cv2.medianBlur(gray,11)

plt.subplot(121)
plt.imshow(gray,cmap='gray')
plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122)
plt.imshow(median,cmap='gray')
plt.title('Blurred')
plt.xticks([])
plt.yticks([])
plt.show()
```

```python
# bilateralFilter
import cv2
import numpy as np
from matplotlib import pyplot as plt
img = cv2.imread('sun.jpg')

image = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)

blur = cv2.bilateralFilter(gray,5,0,0)

plt.subplot(121)
plt.imshow(gray,cmap='gray')
plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122)
plt.imshow(blur,cmap='gray')
plt.title('Blurred')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python

import cv2
import numpy as np
from matplotlib import pyplot as plt
img = cv2.imread('sun.jpg')

image = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
sobel_y = 1/4*np.array([[-1,2,-1],
                   [ 0, 1, 0],
                   [ -1, 2, 1]])


# Filter the image using filter2D, which has inputs: (grayscale image, bit-depth, kernel)
filtered_image = cv2.filter2D(gray, -1, sobel_y)# aafai create gareko kernel filter2d ma aafai banako kernel -1 bhaneeko aafai bannako kernel hai bhaneko
plt.imshow(filtered_image,cmap='gray')
```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('sun.jpg',1)
image = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
edges = cv2.Canny(img,200,250) # high pass filter  200 dekhi 220 ma jada color gradient intensithy mana 

plt.subplot(121),plt.imshow(img,cmap = 'gray')
plt.title('Original Image'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(edges,cmap = 'gray')
plt.title('Edge Image'), plt.xticks([]), plt.yticks([])
plt.show()
```

```python
# Creating a Filter, Edge Detection
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import cv2
import numpy as np
# Reading a image
image = mpimg.imread('sun.jpg')
# plt.imshow(image)
gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
# plt.imshow(gray, cmap='gray')
# Create a custom kernel
# 3x3 array for edge detection
sobel_y = np.array([[0,0,0],
                   [ 0, -4, 0],
                   [ 1, 2, 1]])


# Filter the image using filter2D, which has inputs: (grayscale image, bit-depth, kernel)
filtered_image = cv2.filter2D(gray, -1, sobel_y)

plt.imshow(filtered_image, cmap='gray')
```

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

%matplotlib inline

# Read in the image
image = cv2.imread('dog.jpeg')

# Change color to RGB (from BGR)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Convert the image to grayscale for processing
gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)

plt.imshow(gray, cmap='gray')
wide = cv2.Canny(gray, 30, 100)
tight = cv2.Canny(gray, 160, 240)


# Display the images
f, (ax1, ax2) = plt.subplots(1, 2, figsize=(20,10))

ax1.set_title('wide')
ax1.imshow(wide, cmap='gray')

ax2.set_title('tight')
ax2.imshow(tight, cmap='gray')
```

```python
import cv2
import numpy as np

image = cv2.imread('aayush.jpg')
height, width = image.shape[:2]
print(height,width)

# Let's get the starting pixel coordiantes (top  left of cropping rectangle)
start_row, start_col = int(height * .25), int(width * .25)

# Let's get the ending pixel coordinates (bottom right)
end_row, end_col = int(height * .75), int(width * .75)

# Simply use indexing to crop out the rectangle we desire
cropped = image[start_row:end_row , start_col:end_col]

cv2.imshow("Original Image", image)
cv2.waitKey(0)
cv2.imshow("Cropped Image", cropped)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
