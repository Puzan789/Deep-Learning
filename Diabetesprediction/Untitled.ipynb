```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv('diabetes.csv')
```

```python
df
```

```python
x=df.iloc[:,0:-1]
y=df.iloc[:,-1:]
x
```

```python
x
```

```python
y
```

```python
x.shape
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=3,test_size=0.20)
```

```python
model=Sequential()
model.add(Dense(200,input_dim=8,activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(1, activation='softmax'))
```

```python
model.compile(optimizer='sgd', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
training = model.fit(x_train, y_train, epochs = 200)
```

```python
training.history
```

```python

# list all data in history
print(training.history.keys())
tx = [i for i in range(1,201)]
ty = training.history['accuracy']
# summarize history for accuracy
plt.plot(tx,ty)
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.show()

```

```python
#summarize history for loss
tx = [i for i in range(1,201)]
ty = training.history['loss']
plt.plot(tx,ty)

plt.title('model loss')
plt.legend(['loss', 'epoch'])
plt.ylabel('loss')
plt.xlabel('epoch')
plt.show()
```

```python


test = model.evaluate(x_test,y_test) # gives loss and accuracy
print(test)
```
