```python
import tensorflow as tf
import pandas as pd
import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.linspace(1,100,200)
x
```

```python
# normalizing
x=np.linspace(1,100,200)/100
y=2 * x
x
```

```python
model=Sequential()
model.add(Dense(101,input_dim=1,activation='linear'))
model.add(Dense(60,activation="linear"))
model.add(Dense(35,activation="linear"))
model.add(Dense(25,activation="linear"))
model.add(Dense(15,activation="linear"))
model.add(Dense(5,activation="linear"))
model.add(Dense(3,activation="linear"))
model.add(Dense(2,activation="linear"))
model.add(Dense(1))
model.summary()



```

```python
model.compile('sgd','mse',metrics=['acc'])
```

```python
training = model.fit(x, y, epochs = 100)
```

```python
training.history
```

```python

# list all data in history
print(training.history.keys())
tx = [i for i in range(1,101)]
ty = training.history['acc']
# summarize history for accuracy
plt.plot(tx,ty)
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.show()

```

```python
#summarize history for loss
tx = [i for i in range(1,101)]
ty = training.history['loss']
plt.plot(tx,ty)

plt.title('model loss')
plt.legend(['loss', 'epoch'])
plt.ylabel('loss')
plt.xlabel('epoch')
plt.show()
```

```python
test_x = np.array([1000, 100, 1200, 1900]) / 1900
test_y = 2 * test_x
# test_y = keras.utils.to_categorical(test_y, num_classes=2)

test = model.evaluate(test_x, test_y) # gives loss and accuracy
print(test)
```
